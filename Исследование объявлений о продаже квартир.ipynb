```python
#–°—Ç—Ä–æ–º –æ—Å—å, –ø–æ df. –£–∫–∞–∑—ã–≤–∞–µ–º, —á—Ç–æ–±—ã –≤ df –æ—Å—Ç–∞–ª–∏—Å—å —Ç–æ–ª—å–∫–æ —Å—Ç—Ä–æ–∫–∏ —Å –ø—É—Å—Ç—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ locality_name.
#–°—Ç—Ä–æ–∏–º –ø–æ —ç—Ç–∏–º –¥–∞–Ω–Ω—ã–º –≥—Ä–∞—Ñ–∏–∫ —Ç–∏–ø–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞. –¢–∏–ø - —Å—Ç—É–ø–µ–Ω—á–∞—Ç–∞—è. –ö–æ–ª–∏—á–µ—Å—Ç–æ –∫–æ—Ä–∑–∏–Ω = 50. 
ax = df[df['locality_name'].isnull()].plot(kind='hist', y='last_price', histtype='step', bins=50, range=(0, 45000000), label='delete_locality_name')

#–¢–æ–∂–µ —Å–∞–º–æ–µ –¥–µ–ª–∞–µ–º —Å–æ —Å—Ç–æ–ª–±—Ü–æ–º floors_total. –ü–ª—é—Å –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –¥–æ–±–∞–≤–ª—è–µ–º —Ä–∞–Ω–µ–µ —Å–æ–∑–¥–∞–Ω–Ω—É—é –æ—Å—å
df[df['floors_total'].isnull()].plot(kind='hist', y='last_price', histtype='step', bins=50, label='delete_floors_total', ax=ax, range=(0, 45000000))

#–°—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –ø–æ –Ω–µ–æ—á–∏—â–µ–Ω–Ω–æ–º—É –æ—Ç –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π df
df.plot(kind='hist', y='last_price', range=(0, 45000000),bins=50, label='–°—Ç–æ–∏–º–æ—Å—Ç—å –∫–≤–∞—Ä—Ç–∏—Ä—ã')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ü–µ–Ω—ã –∫–≤–∞—Ä—Ç–∏—Ä—ã')
plt.xlabel('–¶–µ–Ω–∞')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–Ω–æ—Å—Ç—å')
plt.show()

#–£–¥–∞–ª–µ–Ω–∏–µ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ floors_total –∏ locality_name. 
df.dropna(subset=['floors_total', 'locality_name'], inplace=True)
```

#### –ü—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–µ –≤—ã–≤–æ–¥—ã

–£–¥–∞–ª–µ–Ω–∏–µ —Å—Ç—Ä–æ—á–µ–∫ —Å –ø—É—Å—Ç—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö floors_total –∏ locality_name —Å–∏–ª—å–Ω–æ –Ω–µ –ø–æ–≤–ª–∏—è—é—Ç –Ω–∞ –æ–±—â–µ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª—é—á–µ–≤–æ–≥–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è.

### –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö

```python
#0. –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ is_apartment
df['is_apartment'] = df['is_apartment'].fillna(False)

#1. –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ balcony
df['balcony'] = df['balcony'].fillna(0)

#2. –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ days_exposition
df['days_exposition'] = df['days_exposition'].fillna(0)

#3. –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ celing_height –º–µ–¥–∏–∞–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –ø–æ —Å—Ç–æ–ª–±—Ü—É
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())

#4. –ó–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö living_area –∏ kitchen_area —Å –ø–æ–º–æ—â—å—é –º–µ–¥–∏–∞–Ω–Ω–æ–≥–æ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞. 


#–°–æ–∑–¥–∞–¥–∏–º —Å—Ç–æ–ª–±—Ü—ã —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∂–∏–ª–æ–π –∏ –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏, –∞ —Ç–∞–∫–∂–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏ –∫ –æ–±—â–µ–π.
df['liv_coef'] = df['living_area'] / df['total_area']
df['kit_coef'] = df['kitchen_area'] / df['total_area']

#–°–æ–∑–¥–∞–¥–∏–º —Å–ª–æ–≤–∞—Ä–∏
liv_coef_dict = df.groupby('rooms')['liv_coef'].median().to_dict()
kit_coef_dict = df.groupby('rooms')['kit_coef'].median().to_dict()

#–ó–∞–ø–æ–ª–Ω—è–µ–º –ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö –ø–µ—Ä–µ–º–Ω–æ–∂–µ–Ω–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ 
df['living_area'] = (df['rooms'].map(liv_coef_dict))*df['total_area']
df['kitchen_area'] = (df['rooms'].map(kit_coef_dict))* df['total_area']

#–ó–∞–ø–æ–ª–Ω—è–µ–º –æ—Å—Ç–∞–≤—à–∏–µ—Å—è –ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –Ω—É–ª–µ–º. 
df['kitchen_area'] = df['kitchen_area'].fillna(0)

pd.DataFrame(round((df.isna().mean()*100),2)).style.background_gradient('PuBu', low=0.1, high=0.9)
```

### –ü—Ä–∏–≤–æ–¥ –¥–∞–Ω–Ω—ã—Ö –∫ –Ω—É–∂–Ω—ã–º —Ç–∏–ø–∞–º

```python
#0. –î–æ—Ä–∞–±–æ—Ç–∫–∞ –∑–Ω–∞—á–µ–Ω–∏–π –¥–∞—Ç—ã –∏ –≤—Ä–µ–º–µ–Ω–∏
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')

#1. –°–º–µ–Ω–∞ —Ñ–æ—Ä–º–∞—Ç–∞ –¥–∞–Ω–Ω—ã—Ö –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö  floors_total –∏ balcony —Å —Ç–∏–ø–æ–º float

df[['floors_total','balcony']] = df[['floors_total','balcony']].astype('int')

#2. –û—Ç—Å–µ—á–µ–Ω–∏–µ –ª–∏—à–Ω–∏—Ö –¥–∞–Ω–Ω—ã—Ö –≤ —Å—Ç–æ–ª–±—Ü–µ locality_name –ø–æ —Ç–∏–ø—É '–ø–æ—Å–µ–ª–æ–∫, –¥–µ—Ä–µ–≤–Ω—è'
#df['locality_name'].unique()

#–°–æ–∑–¥–∞—ë–º —Ñ—É–Ω—Ü–∏—é. –ù–∞ –≤—Ö–æ–¥ –æ–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Å—Ç—Ä–æ–∫—É. –ü—Ä–æ—Ö–æ–¥–∏—Ç –ø–æ –∫–∞–∂–¥–æ–π –±—É–∫–≤–µ –∏ –∫–æ–≥–¥–∞ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç —Ä–µ–≥–∏—Å—Ç
#–≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—Ç—Ä–æ–∫—É –Ω–∞—á–∏–Ω–∞—è —Å —ç—Ç–æ–π –±—É–∫–≤—ã –∏ –¥–æ –∫–æ–Ω—Ü–∞ —Å—Ç—Ä–æ–∫–∏
def no_upper(string):
    for i in range(len(string)):
        if string[i].isupper():
            return string[i: len(string)]
df['locality_name'] = df['locality_name'].apply(no_upper)

#–ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–∞–Ω–Ω—ã—Ö
#display(df.sample(15,random_state=1))
df.info()
```

# –í—ã–≤–æ–¥—ã –ø–æ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ –¥–∞–Ω–Ω—ã—Ö

1. –ü—Ä–æ–ø—É—Å–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö –∑–∞–ø–æ–ª–Ω–µ–Ω—ã
2. –¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö —É—Å–ø–µ—à–Ω–æ –∏–∑–º–µ–Ω–µ–Ω—ã
3. –ó–Ω–∞—á–µ–Ω–∏–µ –¥–∞—Ç—ã –∏ –≤—Ä–µ–º–µ–Ω–∏ –¥–æ—Ä–∞–±–æ—Ç–∞–Ω—ã
4. –ù–∞–∑–≤–∞–Ω–∏—è –≥–æ—Ä–æ–¥–æ–≤ –∏ –ø–æ—Å–µ–ª–∫–æ–≤ —Å–∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä–æ–≤–∞–Ω—ã

## –†–∞—Å—á—ë—Ç—ã –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –≤ —Ç–∞–±–ª–∏—Ü—É

```python
#0. –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–æ–≤. 
#–¶–µ–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞
df['square_meter_price'] = df['last_price']/ df['total_area']

#–î–æ–±–∞–≤–ª–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤: –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏, –º–µ—Å—è—Ü –∏ –≥–æ–¥ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è;
df['weekday_of_exposition'] = df['first_day_exposition'].dt.weekday
df['month_of_exposition'] = df['first_day_exposition'].dt.month
df['year_of_exposition'] = df['first_day_exposition'].dt.year

#1. –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã: –≤–∞—Ä–∏–∞–Ω—Ç—ã ‚Äî –ø–µ—Ä–≤—ã–π, –ø–æ—Å–ª–µ–¥–Ω–∏–π, –¥—Ä—É–≥–æ–π;
def category_floor(column): 
    floor = column['floor'] #–°–æ–∑–¥–∞–µ–º –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é floor –≤ —Ñ—É–Ω–∫—Ü–∏–∏, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –∏—Å–∫–∞—Ç—å —Å—Ç–æ–ª–±–µ—Ü –≤ floor –≤ df –∏ —Å–æ—Ö—Ä–∞–Ω—è—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ
    floor_total = column['floors_total'] #–¢–∞–∫–∞—è –∂–µ –∏—Å—Ç–æ—Ä–∏—è —Å–æ —Å—Ç–æ–ª–±—Ü–æ–≤ floors_total
    if floor == 1:
        return '–ø–µ—Ä–≤—ã–π'
    elif floor == floor_total:
        return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
    return '–¥—Ä—É–≥–æ–π'
df['floor_category'] = df.apply(category_floor, axis=1) #–ü—Ä–∏–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –∫ df. –£–∫–∞–∑—ã–≤–∞–µ–º, —á—Ç–æ–±—ã —Ä–∞–±–æ—Ç–∞–ª–∞ —Å–æ —Å—Ç–æ–ª–±—Ü–∞–º–∏ axis=1

#2. –°—Ç–æ–ª–±—Ü—ã —Å —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ–º –∂–∏–ª–æ–π –∏ –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏, –∞ —Ç–∞–∫–∂–µ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ–º –ø–ª–æ—â–∞–¥–∏ –∫—É—Ö–Ω–∏ –∫ –æ–±—â–µ–π —É–∂–µ –¥–æ–±–∞–≤–ª–µ–Ω—ã –≤ —Ä–∞–∑–¥–µ–ª–µ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞

df.head()
```

## –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö

### –ò–∑—É—á–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã: –ø–ª–æ—â–∞–¥—å, —Ü–µ–Ω–∞, —á–∏—Å–ª–æ –∫–æ–º–Ω–∞—Ç, –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–æ–≤. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞.


```python
#1. –°—Ç—Ä–æ–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –æ–ø—Ä–µ–¥–µ–ª—è—Ç—å —á–∏—Å–ª–æ –∫–æ—Ä–∑–∏–Ω –≤ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ. 
def my_bins(df,value):
    
    if df[value].max() > 100: #–ï—Å–ª–∏ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —Å—Ç–æ–ª–±—Ü—É –±–æ–ª—å—à–µ 100, —Ç–æ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º 50 –∫–æ—Ä–∑–∏–Ω
        return 50
    elif df[value].max() > 500: #–ï—Å–ª–∏ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —Å—Ç–æ–ª–±—Ü—É –±–æ–ª—å—à–µ 500, —Ç–æ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º 100 –∫–æ—Ä–∑–∏–Ω
        return  100
    elif df[value].max() > 1000: #–ï—Å–ª–∏ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —Å—Ç–æ–ª–±—Ü—É –±–æ–ª—å—à–µ 1000, —Ç–æ –≤–æ–∑–≤—Ä–∞—â–∞–µ–º 200 –∫–æ—Ä–∑–∏–Ω
        return  200
    else: return 20

#2. –°—Ç—Ä–æ–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç —Å—Ç—Ä–æ–∏—Ç—å –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –ø–æ —Å—Ç–æ–ª–±—Ü–∞–º –∏–∑ —Å–ø–∏—Å–∫–∞
def hist_list(df,col):
    #–°—Ç—Ä–æ–∏–º —Å—Ç–æ–ª–±–µ—Ü, –æ—á–∏—â–µ–Ω–Ω—ã–π –æ—Ç –≤—ã–±—Ä–æ—Å–æ–≤ (–æ—Ç 0.05 –∫–≤–∞–Ω—Ç–∏–ª—è –¥–æ 0.95 –∫–≤–∞–Ω—Ç–∏–ª—è. –¢–æ –µ—Å—Ç—å –∏—Å–∫–ª—é—á–∞–µ–º 5% –¥–∞–Ω–Ω—ã—Ö —Å –¥–≤—É—Ö —Å—Ç–æ—Ä–æ–Ω)
    X_1 = df.loc[(df[col] >= df[col].quantile(.05))&(df[col] <= df[col].quantile(.95)),col]
    plt.figure(figsize =(6, 4))
    #–°—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –ø–æ –Ω–∞—à–µ–º—É —Å—Ç–æ–ª–±—Ü—É X_1
    hist = X_1.plot(
                kind='hist',
                bins=my_bins(df, col), 
                alpha=0.5, 
                ec='black',
                color='blueviolet',
                )
    plt.axvline(x=X_1.median(), #–î–æ–±–∞–≤–ª–µ–Ω–∏–µ –ª–∏–Ω–∏–π –Ω–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É. –ì–æ–≤–æ—Ä–∏–º, –ø–æ–∫–∞–∂–∏ –Ω–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ –º–µ–¥–∏–∞–Ω—É –ø–æ —Å—Ç–æ–ª–±—Ü—É
                linewidth=3, #–®–∏—Ä–∏–Ω—É –ª–∏–Ω–∏–∏ —Å–¥–µ–ª–∞–π - 3
                color='green', # –¶–≤–µ—Ç –∑–µ–ª–µ–Ω—ã–π
                linestyle='--') #–°—Ç–∏–ª—å –ª–∏–Ω–∏–∏
    plt.axvline(x=X_1.quantile(.05),
                linewidth=3, 
                color='black', 
                linestyle='--')
    plt.axvline(x=X_1.quantile(.95),
                linewidth=3, 
                color='black', 
                linestyle='--')
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Å—Ç–æ–ª–±—Ü—É ' + col)
    plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
    plt.grid()
    plt.show()
    print(df[col].describe())
    print('--------------------------------------------------------------')
#–°—Ç—Ä–æ–∏–º —Ü–∏–∫–ª, —á—Ç–æ–±—ã —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∏—Ä–æ–≤–∞—Ç—å –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º
list_of_column = ['total_area', 'last_price', 'rooms', 'ceiling_height']
for column in list_of_column:
    hist_list(df, column)
    
```

```python
#–ò–∑—É—á–∞–µ–º –≤—Ä–µ–º—è –ø—Ä–æ–¥–∞–∂–∏. 
X_2 = df['days_exposition'] #–°–æ—Ö—Ä–∞–Ω—è–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø—Ä–æ–¥–∞–∂–∏ –∫–≤–∞—Ä—Ç–∏—Ä—ã –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é.

#–°—Ç—Ä–æ–∏–º –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É
hist = X_2.plot(
    kind='hist', 
    bins=my_bins(df, 'days_exposition'),
    alpha=0.5,
    ec='black',
    color='blueviolet',
)
#–î–æ–±–∞–≤–ª—è–µ–º –ª–∏–Ω–∏–∏ —Å –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º 25%, 50% –∏ 75% –∫–≤–∞—Ä—Ç–∏—Ä–∏–ª—è. 
plt.axvline(x=X_2.median(),
            linewidth=3, 
            color='green', 
            linestyle='--')    
plt.axvline(x=X_2.quantile(.25),
            linewidth=3, 
            color='black', 
            linestyle='--')
plt.axvline(x=X_2.quantile(.75),
            linewidth=3, 
            color='black', 
            linestyle='--')

#–£–∫–∞–∑—ã–≤–∞–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –∏ –Ω–∞–∑–≤–∞–Ω–∏–µ –æ—Å–µ–π.
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Å—Ç–æ–ª–±—Ü—É ' + 'days_exposition')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.grid()
plt.show()

#–°—Ç—Ä–æ–∏–º boxplot –ø–æ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø—Ä–æ–¥–∞–∂–∏ –∫–≤–∞—Ä—Ç–∏—Ä—ã
plt.figure(figsize =(6, 2))
plt.boxplot(X_2,
            notch=True, 
            sym="o", 
            vert=False,
            widths = 0.1, 
            patch_artist = True)
plt.ylabel('–Ø—â–∏–µ —Å —É—Å–∞–º–∏ ' + 'days_exposition')
plt.tick_params(axis = 'both', 
                which = 'major',
                direction = 'inout',
                length = 20,
                width = 4,
                color = 'blueviolet',
                pad = 10,
                labelsize = 15,
                labelcolor = 'black') 
plt.grid()
plt.show()

display(X_2.describe())
print('--------------------------------------------------------------')

#–£–¥–∞–ª–∏–º –≤—ã–±—Ä–æ—Å—ã –≤ –¥–∞–Ω–Ω—ã—Ö (–≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è, –∫–æ—Ç–æ—Ä—ã–µ –ø—Ä–µ–≤—ã—à–∞—é—Ç —Å—É–º–º—É —Ç—Ä–µ—Ç—å–µ–≥–æ –∫–≤–∞—Ä—Ç–∏–ª—è –∏ –ø–æ–ª—Ç–æ—Ä–∞ –∏–Ω—Ç–µ—Ä–∫–≤–∞—Ä—Ç–∞–ª—å–Ω–æ–≥–æ —Ä–∞–∑–º–∞—Ö–∞).
interquartile_range = (df['days_exposition'].quantile(.75) - df['days_exposition'].quantile(.25)) * 1.5
good_days_exposition = df.loc[df['days_exposition'] <= interquartile_range, 'days_exposition']

#–°—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –ø–æ –¥–∞–Ω–Ω—ã–º, –æ—á–∏—â–µ–Ω–Ω—ã—Ö –æ—Ç –≤—ã–±—Ä–æ—Å–æ–≤. 
hist = good_days_exposition.plot(
    kind='hist', 
    bins=100, 
    alpha=0.5, 
    ec='black',
    color='blueviolet',
)
#–î–æ–±–∞–≤–ª—è–µ–º –Ω–∞ –¥–∞–Ω–Ω—ã–µ –ª–∏–Ω–∏–∏ —Å 25%, 50%, 75% –∫–≤–∞—Ä—Ç–∏–ª—è–º–∏. 
plt.axvline(x=good_days_exposition.median(),
            linewidth=3, 
            color='green', 
            linestyle='--')    
plt.axvline(x=good_days_exposition.quantile(.25),
            linewidth=3, 
            color='black', 
            linestyle='--')
plt.axvline(x=good_days_exposition.quantile(.75),
            linewidth=3, 
            color='black', 
            linestyle='--')  
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Å—Ç–æ–ª–±—Ü—É ' + 'good_days_exposition')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.grid()
plt.show()

#–°—Ç—Ä–æ–∏–º boxplot –ø–æ –æ—á–∏—â–µ–Ω–Ω—ã–º –¥–∞–Ω–Ω—ã–º
plt.figure(figsize =(6, 2))
plt.boxplot(good_days_exposition,
            notch=True, 
            sym="o", 
            vert=False,
            widths = 0.1, 
            patch_artist = True)
plt.ylabel('–Ø—â–∏–µ —Å —É—Å–∞–º–∏ ' + 'good_days_exposition')
plt.tick_params(axis = 'both', 
                which = 'major',
                direction = 'inout',
                length = 20,
                width = 4,
                color = 'blueviolet',
                pad = 10,
                labelsize = 15,
                labelcolor = 'black' 
             ) 
plt.grid()
plt.show()

display(good_days_exposition.describe())

#–ù–∞—Ö–æ–¥–∏–º –≤—ã–±—Ä–æ—Å—ã –≤ –æ—á–∏—â–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö 
good_interquartile_range = (good_days_exposition.quantile(.75) - good_days_exposition.quantile(.25)) * 1.5
good_interquartile_range
```

### –í—ã–±–µ—Ä–∏—Ç–µ 10 –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –æ–±—ä—è–≤–ª–µ–Ω–∏–π. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω—é—é —Ü–µ–Ω—É –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –≤ —ç—Ç–∏—Ö –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–∞—Ö. –í—ã–¥–µ–ª–∏—Ç–µ —Å—Ä–µ–¥–∏ –Ω–∏—Ö –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã–µ –ø—É–Ω–∫—Ç—ã —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π –∏ –Ω–∏–∑–∫–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –∂–∏–ª—å—è. –≠—Ç–∏ –¥–∞–Ω–Ω—ã–µ –º–æ–∂–Ω–æ –Ω–∞–π—Ç–∏ –ø–æ –∏–º–µ–Ω–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ 'locality_name'.

```python
#–í—ã–±–µ—Ä–∞–µ–º10 –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º —á–∏—Å–ª–æ–º –æ–±—ä—è–≤–ª–µ–Ω–∏–π. –°—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω—é—é —Ü–µ–Ω—É –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –≤ —ç—Ç–∏—Ö –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–∞—Ö
#–°–æ–∑–¥–∞—ë–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É
pivot_squere = df.pivot_table(index='locality_name', values='square_meter_price', aggfunc=['count', 'mean'])

#–î–∞—ë–º –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤, —á—Ç–æ–±—ã –∏–∑–±–∞–≤–∏—Ç—å—Å—è –æ—Ç –º—É–ª—Ç–∏–∏–Ω–¥–µ–∫—Å–∞
pivot_squere.columns = ['square_meter_price_count', 'square_meter_price_mean']

#–°–æ—Ä—Ç–∏—Ä—É–µ–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É
pivot_squere.sort_values('square_meter_price_count', ascending=False, inplace=True)

#–°–æ—Ö—Ä–∞–Ω—è–µ–º —Ç–æ–ø-10 –∑–Ω–∞—á–µ–Ω–∏–π
top_10 = pivot_squere.head(10).reset_index()

print('')
print('–°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ –≤ –¢–û–ü-10 –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–∞—Ö –ø–æ —á–∏—Å–ª—É –æ–±—ä—è–≤–ª–µ–Ω–∏–π')
display(top_10)


#–°–æ–∑–¥–∞—ë–º –Ω–æ–≤—ã–π df, –∫–æ—Ç–æ—Ä—ã–π –æ–±—ä–µ–¥–∏–Ω—è–µ—Ç –¥–≤–µ —Ç–∞–±–ª–∏—Ü—ã: —Ç–æ–ø-10 –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ –ø–æ —á–∏—Å–ª—É –æ–±—ä—è–≤–ª–µ–Ω–∏–π + –æ–±—â–∏–π df
#–ò—Å–ø–æ–ª—å–∑—É–µ–º left, —á—Ç–æ–±—ã –æ—Å—Ç–∞–≤–∏—Ç—å —Ç–æ–ª—å–∫–æ —Ç–µ –≥–æ—Ä–æ–¥–∞, –∫–æ—Ç–æ—Ä—ã–µ –≤—Ö–æ–¥—è—Ç –≤ –Ω–∞—à–∏ —Ç–æ–ø-10 –≥–æ—Ä–æ–¥–æ–≤ –ø–æ —á–∏—Å–ª—É –æ–±—ä—è–≤–ª–µ–Ω–∏–π
top_10_and_df = top_10.merge(df, on='locality_name', how='left')

#–£–¥–∞–ª—è–µ–º —Å—Ç–∞—Ä—ã–µ —Å—Ç–æ–ª–±—Ü—ã
top_10_and_df = top_10_and_df.drop(['square_meter_price_count', 'square_meter_price_mean'], 1)

#–°—Ç—Ä–æ–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –ø–æ –Ω–æ–≤–æ–π —Ç–∞–±–ª–∏—Ü–µ
top_10_and_df_price = top_10_and_df.pivot_table(index='locality_name', values='last_price', aggfunc=['count','median'])

#–ú–µ–Ω—è–µ–º –Ω–∞–∑–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–æ–≤
top_10_and_df_price.columns = ['last_price_count', 'last_price_median']

#–°–æ—Ä—Ç–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –º–µ–¥–∏–∞–Ω–Ω–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –∫–≤–∞—Ä—Ç–∏—Ä—ã
top_10_and_df_price.sort_values('last_price_median', ascending=False, inplace=True)
print('')
print('–ù–∞—Å–µ–ª—ë–Ω–Ω—ã–µ –ø—É–Ω–∫—Ç—ã —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π –∏ –Ω–∏–∑–∫–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –∂–∏–ª—å—è')
display(top_10_and_df_price.reset_index())
```

##### –ü—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–µ –≤—ã–≤–æ–¥—ã

–ó–∞–¥–∞—á–∞:  –í—ã–¥–µ–ª–∏—Ç–µ —Å—Ä–µ–¥–∏ –Ω–∏—Ö –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã–µ –ø—É–Ω–∫—Ç—ã —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π –∏ –Ω–∏–∑–∫–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –∂–∏–ª—å—è.


–û—Ç–≤–µ—Ç:
–ù–∞—Å–µ–ª–µ–Ω–Ω—ã–µ –ø—É–Ω–∫—Ç—ã —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –∂–∏–ª—å—è: –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥, –ü—É—à–∫–∏–Ω, –ü–∞—Ä–≥–æ–ª–æ–≤–æ.
–ù–∞—Å–µ–ª–µ–Ω–Ω—ã–µ –ø—É–Ω–∫—Ç—ã —Å —Å–∞–º–æ–π –Ω–∏–∑–∫–æ–π —Å—Ç–æ–∏–º–æ—Å—Ç—å—é –∂–∏–ª—å—è: –ú—É—Ä–∏–Ω–æ, –ì–∞—Ç—á–∏–Ω–∞, –í—ã–±–æ—Ä–≥. 

### –ò–∑—É—á–∏—Ç–µ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è –∫–≤–∞—Ä—Ç–∏—Ä: –¥–ª—è –∫–∞–∂–¥–æ–π –∫–≤–∞—Ä—Ç–∏—Ä—ã –µ—Å—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–∏ –¥–æ —Ü–µ–Ω—Ç—Ä–∞. –í—ã–¥–µ–ª–∏—Ç–µ –∫–≤–∞—Ä—Ç–∏—Ä—ã –≤ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ ('locality_name'). –í–∞—à–∞ –∑–∞–¥–∞—á–∞ ‚Äî –≤—ã—è—Å–Ω–∏—Ç—å, –∫–∞–∫–∞—è –æ–±–ª–∞—Å—Ç—å –≤—Ö–æ–¥–∏—Ç –≤ —Ü–µ–Ω—Ç—Ä. –°–æ–∑–¥–∞–π—Ç–µ —Å—Ç–æ–ª–±–µ—Ü —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≤ –∫–∏–ª–æ–º–µ—Ç—Ä–∞—Ö: –æ–∫—Ä—É–≥–ª–∏—Ç–µ –¥–æ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π. –ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω—é—é —Ü–µ–Ω—É –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–∏–ª–æ–º–µ—Ç—Ä–∞. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫: –æ–Ω –¥–æ–ª–∂–µ–Ω –ø–æ–∫–∞–∑—ã–≤–∞—Ç—å, –∫–∞–∫ —Ü–µ–Ω–∞ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —É–¥–∞–ª—ë–Ω–Ω–æ—Å—Ç–∏ –æ—Ç —Ü–µ–Ω—Ç—Ä–∞. –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –≥—Ä–∞–Ω–∏—Ü—É, –≥–¥–µ –≥—Ä–∞—Ñ–∏–∫ —Å–∏–ª—å–Ω–æ –º–µ–Ω—è–µ—Ç—Å—è ‚Äî —ç—Ç–æ –∏ –±—É–¥–µ—Ç —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–∞—è –∑–æ–Ω–∞.

```python
#–°–æ–∑–¥–∞–µ–º –Ω–æ–≤—ã–π df, –≤ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥—É—Ç —Å—Ç—Ä–æ–∫–∏ —Ç–æ–ª—å–∫–æ —Å –°–ü–ë

spb_flat = df.query('locality_name == "–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥"')

#–°–æ–∑–¥–∞—ë–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü - —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≤ –∫–∏–ª–æ–º–µ–Ω—Ç—Ä–∞—Ö 
spb_flat['city_center_km'] = (spb_flat['cityCenters_nearest'] / 1000)

#–£–¥–∞–ª—è–µ–º –ø—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ, —á—Ç–æ–±—ã –æ–∫—Ä—É–≥–ª–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è
spb_flat.dropna(subset=['city_center_km'], inplace=True)

#–û–∫—Ä—É–≥–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –≤ —Å—Ç–æ–ª–±—Ü–µ –¥–æ —Ü–µ–ª–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
spb_flat['city_center_km'] = spb_flat['city_center_km'].astype('int')

#–°—Ç—Ä–æ–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É
spb_flat_price = spb_flat.pivot_table(index='city_center_km', values='last_price')

display(spb_flat_price.head(10))

#–°—Ç—Ä–æ–∏–º —Å—Ç–æ–ª–±—á–∞—Ç—ã–π –≥—Ä–∞—Ñ–∏–∫
spb_flat_price.plot(kind='bar', grid=True, color='blueviolet')
plt.title('–¶–µ–Ω—ã –∫–≤–∞—Ä—Ç–∏—Ä—ã –æ—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –¥–æ —Ü–µ–Ω—Ç—Ä–∞')
plt.xlabel('–†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –æ—Ç —Ü–µ–Ω—Ç—Ä–∞ –≤ –∫–º')
plt.ylabel('last_price')
plt.show()
```
