```python
def hist_boxplot(data, hue):
    '''–≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è —Å—Ç—Ä–æ–∏—Ç –¥–≤–∞ –≥—Ä–∞—Ñ–∏–∫–∞ histplot –∏ boxplot.
    –ù–∞ –≤—Ö–æ–¥ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–∏–º–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏ –∏ —Å—Ç–æ–ª–±–µ—Ü –ø–æ –∫–æ—Ç–æ—Ä–æ–º—É –∏–¥—ë—Ç —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π.
    '''
    for column in data.drop(hue, axis=1).columns:
        fig, ax = plt.subplots(2,1, figsize=(16,10))
        sns.histplot(data=data, x=column, hue=hue, kde=True, ax=ax[0])
        sns.boxplot(data=data, x=column, y=hue, hue=hue, ax=ax[1], orient = 'h')
        ax[0].set_xlabel('–ó–Ω–∞—á–µ–Ω–∏—è –ø–æ ' + column, fontsize=15)
        ax[1].set_xlabel('–ó–Ω–∞—á–µ–Ω–∏—è –ø–æ ' + column, fontsize=15)
        ax[0].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=15)
        ax[1].set_ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ –ø–æ ' + hue, fontsize=15)
        ax[0].set_title('–ö–æ–ª-–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ' + column, fontsize=20)
        ax[1].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ' + column, fontsize=20)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.subplots_adjust(hspace=0.5)
        plt.show()
        print('–û–ø–∏—Å–∞—Ç–µ–ª—å–Ω—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ' + column)
        display(data.groupby(hue)[column].describe())
```

–í –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π score –Ω–∞–±–ª—é–¥–∞—é—Ç—Å—è –≤—Å–ø–ª–µ—Å–∫–∏ –≤ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö, –∫–∞–∫ —Å—Ä–µ–¥–∏ —É—à–µ–¥—à–∏—Ö, —Ç–∞–∫ –∏ —Å—Ä–µ–¥–∏ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –∫–ª–∏–µ–Ω—Ç–æ–≤. 

–í —Å—Ç–æ–ª–±—Ü–µ age –Ω–∞–±–ª—é–¥–∞—é—Ç—Å—è –ø–∏–∫–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ 29 –∏ 40 –ª–µ—Ç —Å—Ä–µ–¥–∏ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –∫–ª–∏–µ–Ω—Ç–æ–≤. –ü–æ—Å–ª–µ 40 –ª–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —É—à–ª–∏ –∏–∑ –±–∞–Ω–∫–∞ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è, –∞ —Ç–µ—Ö, –∫—Ç–æ –æ—Å—Ç–∞–ª—Å—è - —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è. –ü–æ—Å–ª–µ 50 –ª–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤ –æ–±–µ–∏—Ö –≥—Ä—É–ø–ø–∞—Ö —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –ø—Ä–∏–º–µ—Ä–Ω–æ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–º. 

–°—Ä–µ–¥–Ω–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏: –¥–µ–±–µ—Ç–æ–≤—ã–π –æ—Å—Ç–∞—Ç–æ–∫ –Ω–∞ —Å—á–µ—Ç–µ, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤ –≤ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ—Å—Ç–∏ –∏ —Å—Ä–µ–¥–Ω–∏–π —Ä–∞–∑–º–µ—Ä –¥–æ—Ö–æ–¥–∞ –≤ –¥–≤—É—Ö –≥—Ä—É–ø–ø–∞—Ö –æ—á–µ–Ω—å –±–ª–∏–∑–∫–∏.

**–†–∞—Å—Å—á–∏—Ç–∞–µ–º churn rate –≤ –∫–∞–∂–¥–æ–π –≤–æ–∑—Ä–∞—Å—Ç–Ω–æ–π –≥—Ä—É–ø–ø–µ**

```python
def active_churn_count(column, range_start, range_end, step):
    '''
    –ü—Ä–∏–º–µ—Ä: active_churn_count('age', 18, 90, 5). –ü–æ–ª—É—á–∏–º –ø—Ä–æ—Ü–µ–Ω—Ç –æ—Ç—Ç–æ–∫–∞
    –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –≤–æ–∑—Ä–∞—Å—Ç –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –æ—Ç 18 –¥–æ 90 –ª–µ—Ç —Å —à–∞–≥–æ–º 5 –ª–µ—Ç . '''
    col_from = []
    col_to = []
    number_active_clients = []
    number_churn_clients = []
    churn = df.loc[df['churn'] == 1]
    churn_no = df.loc[df['churn'] == 0]
    
    for number in range(range_start, range_end, step):
        threshold = number - step
        churn_clients = churn.loc[(churn[column] <=number) & (churn[column] >= threshold)][column].count()
        active_clients = churn_no.loc[(churn_no[column]<=number)&(churn_no[column]>= threshold)][column].count()
        col_from.append(number - step)
        col_to.append(number)
        number_churn_clients.append(churn_clients)
        number_active_clients.append(active_clients)
    final_table = pd.DataFrame(columns=['from_col', 'to_col', 'active_clients', 'churn_clients'])
    final_table['from_col'] = col_from
    final_table['from_col'] = final_table['from_col'].astype('object')
    final_table['to_col'] = col_to
    final_table['to_col'] = final_table['to_col'].astype('object')
    final_table['active_clients'] = number_active_clients
    final_table['churn_clients'] = number_churn_clients
    final_table['churn_rate %'] = round((final_table['churn_clients'] / final_table['active_clients']* 100),2)
    final = (final_table.style
                        .background_gradient(cmap='YlOrRd', subset=['churn_rate %'])
                        .bar(subset=['active_clients','churn_clients'], align='mid', color=['#d65f5f','#5fba7d'])
                        .set_caption('–ö–æ–ª-–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –∏ —É—à–µ–¥—à–∏—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π ' + column)
                        .set_properties(**{'text-align': 'center', 'padding': '8px'}))
    return final
```

```python
def metrics_plot(model, features_valid, target_valid, graph_name_model):
        probabilities_valid = model.predict_proba(features_valid)
        precision, recall, thresholds = precision_recall_curve(target_valid, probabilities_valid[:, 1])
        fpr, tpr, thresholds = roc_curve(target_valid, probabilities_valid[:, 1])
            
        fig, ax = plt.subplots(1, 2, figsize=(15, 6))
        fig.subplots_adjust(hspace=0.4, wspace=0.4)
        
        sns.lineplot(recall, precision, drawstyle='steps-post', ax=ax[0])
        ax[0].set_xlabel('Recall')
        ax[0].set_ylabel('Precision')
        ax[0].set_ylim([0.0, 1.05])
        ax[0].set_xlim([0.0, 1.0])
        ax[0].set_title('–ö—Ä–∏–≤–∞—è Precision-Recall –¥–ª—è –º–æ–¥–µ–ª–∏ ' + graph_name_model)
        
        sns.lineplot(fpr, tpr, ax=ax[1])
        ax[1].plot([0, 1], [0, 1], linestyle='--')
        ax[1].set_xlim(0,1)
        ax[1].set_ylim(0,1)
        ax[1].set_xlabel('False Positive Rate')
        ax[1].set_ylabel('True Positive Rate')
        ax[1].set_title('ROC-–∫—Ä–∏–≤–∞—è –¥–ª—è –º–æ–¥–µ–ª–∏ ' +graph_name_model)
```

```python
def cf_matrix_heat(y_test, y_pred):
    cf_matrix = confusion_matrix(y_test, y_pred)
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ['{0:0.0f}'.format(value) for value in cf_matrix.flatten()]
    group_percentages = ['{0:.2%}'.format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]
    labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    fix, ax = plt.subplots(figsize=(17,7))
    sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')
```

```python
churn_df = df.loc[df['churn'] == 1]
no_churn_df = df.loc[df['churn'] == 0]
columns_list = ['estimated_salary', 'age']

def shapiro_test(data, churn_name):
    for column in columns_list:
        stat, p_value = scipy.stats.shapiro(data[column])
        if p_value > alpha:
            print(f'–ü—Ä–∏–Ω–∏–º–∞–µ–º –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–∞ {column} –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ {churn_name}.')
        else:
            print(f'–û—Ç–∫–ª–æ–Ω—è–µ–º –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–∞ {column} –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ {churn_name}.')
```

```python
# –°–æ–∑–¥–∞—ë–º —Ñ—É–Ω–∫—Ü–∏—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
def mean_distribution(column): 
    churn_list = [] 
    no_churn_list = []
    df_churn = df.loc[df['churn']==1]
    df_no_churn = df.loc[df['churn']==0]
    len_churn = len(df_churn) 
    len_no_churn = len(df_no_churn)
    for _ in tqdm(range(10000)):
        sample_churn = np.random.choice(df_churn[column], size=len_churn, replace=True)
        sample_no_churn = np.random.choice(df_no_churn[column], size=len_no_churn, replace=True)
        churn_list.append(sample_churn.mean())
        no_churn_list.append(sample_no_churn.mean())
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,5))
    ax.set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö —Å—Ä–µ–¥–∏ —Ç–µ—Ö, –∫—Ç–æ —É—à—ë–ª –∏ –æ—Å—Ç–∞–ª—Å—è. –§–∏–æ–ª–µ—Ç–æ–≤—ã–π —Ü–≤–µ—Ç - —É—à–ª–∏, –û—Ä–∞–Ω–∂–µ–≤—ã–π - –æ—Å—Ç–∞–ª–∏—Å—å')
    ax.hist(churn_list, bins=100, color='blueviolet', alpha=0.5)
    ax.hist(no_churn_list, bins=100, color='orange', alpha=0.5)
    fig.tight_layout()
```

```python
def mannwhitneyu_test(column):
    stat, p_value = mannwhitneyu(churn_df[column],no_churn_df[column],alternative='two-sided')
    print(f'p-value= {p_value:.2f}')
    print('')
    print('–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Å—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥ —É—à–µ–¥—à–∏—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤ —Ç–∞–∫–æ–π –∂–µ, –∫–∞–∫ –∏ —É –ª–æ—è–ª—å–Ω—ã—Ö –∫ –±–∞–Ω–∫—É –∫–ª–∏–µ–Ω—Ç–æ–≤.')
    print('–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Å—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥ —É—à–µ–¥—à–∏—Ö –∏ –ª–æ—è–ª—å–Ω—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è.')
    print('')
    if (p_value < alpha):
        print(f'–û—Ç–∫–ª–æ–Ω—è–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É {column} –º–µ–∂–¥—É —É—à–µ–¥—à–∏–º–∏ –∏ –æ—Å—Ç–∞–≤—à–∏–º–∏—Å—è –∫–ª–∏–µ–Ω—Ç–∞–º–∏')
    else:
        print(f'–ù–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –¥–ª—è –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –Ω—É–ª–µ–≤–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É {column}')
```
