```python
# Import libraries
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
%matplotlib inline
%config InlineBackend.figure_format='retina'


from datetime import date, time, timedelta
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
from tqdm import tqdm

# Turn off warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
def first_look(df):
    '''Function displays main info about df'''
    print('-------------- –ü–µ—Ä–≤—ã–µ 5 —Å—Ç—Ä–æ–∫ --------------')
    display(df.head())
    print('')
    print('-------------- –¢–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö --------------')
    display(df.info())
    print('')
    print('-------------- –ü—Ä–æ–ø—É—Å–∫–∏ --------------')
    count = 0
    for element in df.columns:
        if df[element].isna().sum() > 0:
            print(element, ' - ', df[element].isna().sum(), '–ø—Ä–æ–ø—É—Å–∫–æ–≤')
            count = +1
    if count == 0:
            print('–ü—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ—Ç')
    print('')
    print('-------------- –î—É–±–ª–∏–∫–∞—Ç—ã --------------')
    if df.duplicated().sum() > 0:
        print('–î—É–±–ª–∏–∫–∞—Ç–æ–≤: ', df.duplicated().sum())
    else:
        print('–î—É–±–ª–∏–∫–∞—Ç–æ–≤ –Ω–µ—Ç')      
```

```python
#Read data
hypothesis = pd.read_csv('/datasets/hypothesis.csv')
hypothesis.columns = hypothesis.columns.str.lower()
first_look(hypothesis)
```

### –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º. –°–¥–µ–ª–∞–π—Ç–µ –≤—ã–≤–æ–¥—ã –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è.

```python
# —Å–æ–∑–¥–∞–¥–∏–º –º–∞—Å—Å–∏–≤ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–∞—Ä –∑–Ω–∞—á–µ–Ω–∏–π –¥–∞—Ç –∏ –≥—Ä—É–ø–ø —Ç–µ—Å—Ç–∞
datesGroups = orders[['date', 'group']].drop_duplicates()

# —Å–æ–±–µ—Ä—ë–º –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–µ –ø–æ –¥–Ω—è–º –¥–∞–Ω–Ω—ã–µ –æ –∑–∞–∫–∞–∑–∞—Ö –∏ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è—ä
ordersAggregated = datesGroups.apply(
    lambda x: orders[
        np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])
    ].agg(
    {
        'date':'max',
        'group':'max',
        'transactionId':pd.Series.nunique,
        'visitorId':pd.Series.nunique,
        'revenue':'sum'
    }
    ),axis=1,
).sort_values(by=['date', 'group'])

visitorsAggregated = datesGroups.apply(
    lambda x: visitors[
        np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])
    ].agg(
    {
        'date':'max',
        'group':'max',
        'visitors':'sum'
    }
    ), axis=1,
).sort_values(by=['date', 'group'])


# –æ–±—ä–µ–¥–∏–Ω–∏–º –æ–±–µ —Ç–∞–±–ª–∏—Ü—ã –≤ –æ–¥–Ω—É —Å –ø–æ–Ω—è—Ç–Ω—ã–º–∏ –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ —Å—Ç–æ–ª–±—Ü–æ–≤
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])

cumulativeData.columns = [
    'date',
    'group',
    'orders',
    'buyers',
    'revenue',
    'visitors',
]
```

```python
# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–∫–∞–∑–æ–≤ –∏ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–æ–π –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ –ê
cumulativeRevenueA = cumulativeData[cumulativeData['group'] == 'A'][['date', 'revenue', 'orders']]
# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–∫–∞–∑–æ–≤ –∏ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–æ–π –ø–æ –¥–Ω—è–º –≤ –≥—Ä—É–ø–ø–µ B
cumulativeRevenueB = cumulativeData[cumulativeData['group'] == 'B'][['date', 'revenue', 'orders']]

# –∑–∞–¥–∞—ë–º —Ä–∞–∑–º–µ—Ä—ã —Ñ–∏–≥—É—Ä—ã
plt.figure(figsize=(15,6))

# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –≤—ã—Ä—É—á–∫–∏ –≥—Ä—É–ø–ø—ã –ê
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label='–í—ã—Ä—É—á–∫–∞ –ø–æ –≥—Ä—É–ø–ø–µ –ê')
#–°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –≤—ã—Ä—É—á–∫–∏ –≥—Ä—É–ø–ø—ã B
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label='–í—ã—Ä—É—á–∫–∞ –ø–æ –≥—Ä—É–ø–ø–µ –í')

# –∑–∞–¥–∞—ë–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≥—Ä–∞—Ñ–∏–∫–∞
plt.legend(fontsize=16)
plt.title('–ì—Ä–∞—Ñ–∏–∫ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º', fontsize=22)
plt.xlabel('–î–∞—Ç–∞', fontsize=16)
plt.ylabel('–ö—É–º—É–ª—è—Ç–∏–≤–Ω–∞—è –≤—ã—Ä—É—á–∫–∞', fontsize=16)
plt.grid(True)
plt.annotate('—Å–∫–∞—á–æ–∫', xy=('2019-08-17', 2800000), xytext=('2019-08-10', 3500000),
            arrowprops=dict(facecolor='g'))
plt.annotate('–æ—Ç—Ä—ã–≤', xy=('2019-08-05', 800000), xytext=('2019-08-05', 2000000),
            arrowprops=dict(facecolor='g')) 
plt.show()

topRevenueDays = orders[orders['group'] == 'B'].groupby('date').agg(
    {'revenue':'sum'}).sort_values(by='revenue', ascending=False)
display(topRevenueDays.head().style.background_gradient(cmap='YlOrRd'))
print('–ú–µ–¥–∏–∞–Ω–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –∑–∞ –ø–µ—Ä–∏–æ–¥ {}'.format(topRevenueDays['revenue'].median()))
```

```python
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']
visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsACumulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <=x['date']].agg(
        {'date':'max', 'visitorsPerDateA':'sum'}
    ), 
    axis=1,
)
visitorsACumulative.columns = ['date', 'visitorsCumulativeA']

visitorsBCumulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
    {'date':'max', 'visitorsPerDateB':'sum'}
    ),
    axis=1,
)
visitorsBCumulative.columns = ['date', 'visitorsCumulativeB']
```

```python
ordersADaily = (
    orders[orders['group'] == 'A'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue':'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersBDaily = (
    orders[orders['group'] == 'B'][['date', 'transactionId', 'visitorId', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transactionId': pd.Series.nunique, 'revenue':'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersACumulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date':'max', 'ordersPerDateA':'sum', 'revenuePerDateA':'sum'}
    ),
    axis=1,
).sort_values(by='date')

ordersACumulative.columns = ['date', 'ordersCumulativeA', 'revenueCumulativeA']

ordersBCumulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date':'max','ordersPerDateB':'sum', 'revenuePerDateB':'sum'}
    ), 
    axis=1,
)

ordersBCumulative.columns = ['date', 'ordersCumulativeB', 'revenueCumulativeB']

data = (
    ordersADaily.merge(ordersBDaily, left_on='date', right_on='date', how='left')
    .merge(ordersACumulative, left_on='date', right_on='date', how='left')
    .merge(ordersBCumulative, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACumulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCumulative, left_on='date', right_on='date', how='left')
)
data.head()
```

–û–±—ä—è–≤–∏–º –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ sampleA –∏ sampleB, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º –∏–∑ —Ä–∞–∑–Ω—ã—Ö –≥—Ä—É–ø–ø –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤. –¢–µ–º, –∫—Ç–æ –Ω–∏—á–µ–≥–æ –Ω–µ –∑–∞–∫–∞–∑–∞–ª, –±—É–¥—É—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –Ω—É–ª–∏. –≠—Ç–æ –Ω—É–∂–Ω–æ, —á—Ç–æ–±—ã –ø–æ–¥–≥–æ—Ç–æ–≤–∏—Ç—å –≤—ã–±–æ—Ä–∫–∏ –∫ –ø—Ä–æ–≤–µ—Ä–∫–µ –∫—Ä–∏—Ç–µ—Ä–∏–µ–º –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏.

1. –ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –¥–≤—É—Ö –≥—Ä—É–ø–ø –ø–æ "—Å—ã—Ä—ã–º" –¥–∞–Ω–Ω—ã–º —Ä–∞–≤–Ω—ã. –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –Ω–µ—Ç. 
2. –û–±—Ä–∞—Ç–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –¥–≤—É—Ö –≥—Ä—É–ø–ø –ø–æ "—Å—ã—Ä—ã–º" –¥–∞–Ω–Ω—ã–º –Ω–µ —Ä–∞–≤–Ω—ã. –ï—Å—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è.

__–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –æ—Ç —Ä–µ–≤—å—é–µ—Ä–∞ ‚Ññ1__

–ù—É–∂–Ω–æ —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –≥–∏–ø–æ—Ç–µ–∑—ã –ø–µ—Ä–µ–¥ –∫–∞–∂–¥—ã–º —Ç–µ—Å—Ç–æ–º, –∏–Ω–∞—á–µ —á—Ç–æ –º—ã –ø—Ä–æ–≤–µ—Ä—è–µ–º? –ü—Ä–∏–º–µ—Ä–Ω–æ —Ç–∞–∫:

2. –û–±—Ä–∞—Ç–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: ...  
    
–¢–∞–∫ —Å—Ç–æ—Ä–æ–Ω–Ω–∏–π –Ω–∞–±–ª—é–¥–∞—Ç–µ–ª—å –ø–æ–π–º–µ—Ç —á—Ç–æ —Ç—ã –ø—Ä–æ–≤–µ—Ä—è–µ—à—å.
    
–ê –≤–æ—Ç –≤ [—ç—Ç–æ–π —Å—Ç–∞—Ç—å–µ](https://ru.qwe.wiki/wiki/Null_hypothesis) –º–æ–∂–Ω–æ –ø–æ—á–∏—Ç–∞—Ç—å –≤ —Ü–µ–ª–æ–º –æ —Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω–∏–∏ –≥–∏–ø–æ—Ç–µ–∑ –∏ –∫–∞–∫ –≤—Å—ë —ç—Ç–æ –¥–µ–ª–æ –Ω–∞—á–∞–ª–æ—Å—å)

üëç

```python
sampleA = pd.concat(
    [
        ordersByUsersA['orders'], 
        pd.Series(0, 
                  index=np.arange(
         data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
                  ), 
                  name='orders'
                 ),
    ],
    axis=0
)

sampleB = pd.concat(
    [
        ordersByUsersB['orders'], 
        pd.Series(0, 
                  index = np.arange (
        data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
                  ), 
                  name='orders'
                 )
    ], 
    axis=0
)

# –ø—Ä–∏–º–µ–Ω–∏–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏:

results = stats.mannwhitneyu(sampleA, sampleB, alternative = 'two-sided')

alpha = 0.05

print('–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –¥–≤—É—Ö –≥—Ä—É–ø–ø –ø–æ "—Å—ã—Ä—ã–º" –¥–∞–Ω–Ω—ã–º —Ä–∞–≤–Ω—ã. –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –Ω–µ—Ç.')
print('–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –¥–≤—É—Ö –≥—Ä—É–ø–ø –ø–æ "—Å—ã—Ä—ã–º" –¥–∞–Ω–Ω—ã–º –Ω–µ —Ä–∞–≤–Ω—ã. –ï—Å—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è.')
print('')
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ: {0:.3f}'.format(results.pvalue))
if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: —Ä–∞–∑–Ω–∏—Ü–∞ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞.')
else:
    print(
        '–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, –≤—ã–≤–æ–¥ –æ —Ä–∞–∑–ª–∏—á–∏–∏ —Å–¥–µ–ª–∞—Ç—å –Ω–µ–ª—å–∑—è'
    ) 
print('–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –≥—Ä—É–ø–ø—ã B —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {0:.2%}.'.format(sampleB.mean() / sampleA.mean() - 1))
```

–ü–µ—Ä–≤–æ–µ —á–∏—Å–ª–æ ‚Äî p-value = 0.011 –º–µ–Ω—å—à–µ 0.05. –ó–Ω–∞—á–∏—Ç, –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ç–æ–º, —á—Ç–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –≤ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –Ω–µ—Ç, –æ—Ç–≤–µ—Ä–≥–∞–µ–º. –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç –≥—Ä—É–ø–ø—ã B —Ä–∞–≤–µ–Ω 15.98% ‚Äî –≤—Ç–æ—Ä–æ–µ —á–∏—Å–ª–æ –≤ –≤—ã–≤–æ–¥–µ.

### –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º. –°–¥–µ–ª–∞–π—Ç–µ –≤—ã–≤–æ–¥—ã –∏ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è.


```python

#–ü—Ä–∏–º–µ–Ω–∏–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∫—Ä–∏—Ç–µ—Ä–∏–π –ú–∞–Ω–Ω–∞-–£–∏—Ç–Ω–∏ –∫ –ø–æ–ª—É—á–µ–Ω–Ω—ã–º –≤—ã–±–æ—Ä–∫–∞–º:

results = stats.mannwhitneyu(
            orders[
                np.logical_and(
                    orders['group'] == 'A',
                    np.logical_not(orders['visitorId'].isin(usersWithExpensiveOrders)),
                )
            ]['revenue'],
            orders[
                np.logical_and(
                    orders['group'] == 'B',
                    np.logical_not(orders['visitorId'].isin(usersWithExpensiveOrders)),
                )
            ]['revenue'],alternative = 'two-sided')
alpha = 0.05

print('–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Å—Ä–µ–¥–Ω–∏–µ —á–µ–∫–∏ –¥–≤—É—Ö –≥—Ä—É–ø–ø –ø–æ "–æ—á–∏—â–µ–Ω–Ω—ã–º" –¥–∞–Ω–Ω—ã–º —Ä–∞–≤–Ω—ã. –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –Ω–µ—Ç.')
print('–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: —Å—Ä–µ–¥–Ω–∏–µ —á–µ–∫–∏ –¥–≤—É—Ö –≥—Ä—É–ø–ø –ø–æ "–æ—á–∏—â–µ–Ω–Ω—ã–º" –¥–∞–Ω–Ω—ã–º –Ω–µ —Ä–∞–≤–Ω—ã. –ï—Å—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è.')
print('')
print('p-–∑–Ω–∞—á–µ–Ω–∏–µ: {0:.3f}'.format(results.pvalue))
if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: —Ä–∞–∑–Ω–∏—Ü–∞ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞.')
else:
    print(
        '–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, –≤—ã–≤–æ–¥ –æ —Ä–∞–∑–ª–∏—á–∏–∏ —Å–¥–µ–ª–∞—Ç—å –Ω–µ–ª—å–∑—è'
    ) 
print(
    '–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≥—Ä—É–ø–ø–µ –ø–æ –æ—á–∏—â–µ–Ω–Ω—ã–º –¥–∞–Ω–Ω—ã–º —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç {0:.3%}'.format(
        orders[
            np.logical_and(
                orders['group'] == 'B',
                np.logical_not(orders['visitorId'].isin(usersWithExpensiveOrders)),
            )
        ]['revenue'].mean()
        / orders[
            np.logical_and(
                orders['group'] == 'A',
                np.logical_not(orders['visitorId'].isin(usersWithExpensiveOrders)),
            )
        ]['revenue'].mean()
        - 1
    )
)
```

```python
usersAB = orders[['visitorId', 'date', 'revenue', 'group']]
usersAB = orders[~orders['visitorId'].isin(usersWithExpensiveOrders)]

def cum_data(x):
    data = usersAB[
        np.logical_and(
            usersAB['date'] <= x['date'],
            usersAB['group'] == x['group']
        )
    ]
    
    data = data.agg(
        {'date' : 'max',
         'group' : 'max',
         'visitorId' : pd.Series.nunique,
         'revenue' : 'sum'
        }
    )
    return data

cumulative = datesGroups.apply(cum_data, axis=1)
cumulative = cumulative.rename(columns={'visitorId':'paid_users'})
cumulative = cumulative.sort_values(by='date')

ab_totals = visitors.groupby('group', as_index=False)['visitors'].sum()
ab_totals = ab_totals.rename(columns={'visitors':'total_users'})
cumulative = cumulative.merge(ab_totals, on='group')
cumulative.head()

cumulative['arpu'] = cumulative['revenue'] / cumulative['total_users']
cumulative['arppu'] = cumulative['revenue'] / cumulative['paid_users']
cumulative['cr'] = cumulative['paid_users'] / cumulative['total_users']

cumulative_a = cumulative[cumulative['group'] == 'A']
cumulative_b = cumulative[cumulative['group'] == 'B']

cumulative_final = cumulative_a.merge(
    cumulative_b,
    on='date',
    suffixes=('_A', '_B')
)
cumulative_final['date'] = pd.to_datetime(cumulative_final['date'])
cumulative_final = cumulative_final.sort_values(by='date')
```
