### –ü–æ—Å–º–æ—Ç—Ä–∏—Ç–µ, –∫–∞–∫ –º–µ–Ω—è–ª–∏—Å—å –ø—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º. –í—ã–±–µ—Ä–∏—Ç–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –∏ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –≥–æ–¥–∞–º. –ó–∞ –∫–∞–∫–æ–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã–π —Å—Ä–æ–∫ –ø–æ—è–≤–ª—è—é—Ç—Å—è –Ω–æ–≤—ã–µ –∏ –∏—Å—á–µ–∑–∞—é—Ç —Å—Ç–∞—Ä—ã–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã?

```python
def sales_region_all_years(data, column, after, before, sales_threshold_on, sales_threshold_sum):
    result = 0
    sample = data.pivot_table(index='platform', 
                              columns='year_of_release', 
                              values=column,aggfunc='sum')
    sample = pd.DataFrame(sample).reset_index()
    sample.fillna(0, inplace=True)
    
    for year in range(after, before):
        result += sample[year]
        
    sample[column] = result
    sample = sample[['platform', column]]
    sample = sample.sort_values(column, ascending=False).head(6)
    display(sample.style.background_gradient(cmap='Blues'))
    
    print('----------------------------------------------------')
    
    sales = data
    sales = sales.loc[(sales[column] >= sales[column].quantile(.03))&
                                    (sales[column] <= sales[column].quantile(.97))]
    sales_platform = sales[(sales['year_of_release'] >after) & 
                           (sales['year_of_release'] <before)]
 
    if sales_threshold_on == 'yes':
        sales_platform = sales_platform.loc[sales_platform[column] > sales_threshold_sum]
    else:
        sales_platform = sales_platform
    ax = sns.catplot(data=sales_platform, 
                     x='year_of_release', 
                     y=column,
                     hue='platform', 
                     kind='box', 
                     aspect=1.5)
    plt.grid(True)
    plt.title(f'–ü—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º —Å {after} –ø–æ {before} –≥–≥', fontsize=14)
    plt.xlabel('–ì–æ–¥', fontsize=12)
    plt.ylabel('–ú–∏—Ä–æ–≤—ã–µ –ø—Ä–æ–¥–∞–∂–∏, –º–ª–Ω. —à—Ç', fontsize=12)
    plt.show()
```

```python
sales_region_all_years(df, 'sales_world',1980, 1990, 'yes', 1)
```

```python
def sales_region_last_5_years(data, column):
    sample = data.loc[data['year_of_release'] > 2011].pivot_table(index='platform', 
                                                                  columns='year_of_release', 
                                                                  values=column,
                                                                  aggfunc='sum')
    sample = pd.DataFrame(sample).reset_index()
    sample.fillna(0, inplace=True)
    sample[column] = sample[2012] + sample[2013] +  sample[2014] + sample[2015] + sample[2016]
    sample = sample.sort_values(column, ascending=False).head(6)
    display(sample.style.background_gradient(cmap='Blues'))
    
    print('---------------------------------------------------------')
    
    top_5 = data.query('platform in(["PS4", "PS3", "X360", "3DS", "XOne"])')
    top_5 = top_5.loc[(top_5[column] >= top_5[column].quantile(.03))&
                                    (top_5[column] <= top_5[column].quantile(.97))]
    top_5_last_5_years = top_5[top_5['year_of_release'] >2011]
    
    ax = sns.catplot(data=top_5_last_5_years, 
                     x='year_of_release', 
                     y=column, 
                     hue='platform', 
                     kind='box',  
                     aspect=2)
    plt.subplots_adjust(0)
    plt.grid(True)
    plt.title(f'–ü—Ä–æ–¥–∞–∂–∏ –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º —Å 2012 –ø–æ 2016 –≥–≥', fontsize=16)
    plt.xlabel('–ì–æ–¥', fontsize=12)
    plt.ylabel('–ú–∏—Ä–æ–≤—ã–µ –ø—Ä–æ–¥–∞–∂–∏, –º–ª–Ω. —à—Ç', fontsize=12)
    plt.show()
```

```python
def cor_plot_and_matrix(data, platform_name):
    platform = data[data['platform'] == platform_name]
    corr_price = platform[['user_score', 'critic_score', 'sales_world']]
    display(platform.corr()['sales_world'].sort_values())
    f, ax = plt.subplots(figsize=(10, 5))
    ax = sns.heatmap(corr_price.corr(), 
                annot = True, 
                vmin=-1, 
                vmax=1, 
                center= 0, 
                cmap= 'coolwarm', 
                linewidths=1, 
                linecolor='black')
    plt.tick_params(axis = 'both', which = 'major', labelsize = 12)
    plt.title('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã ' + platform_name, fontsize=14)
    plt.show()
    pd.plotting.scatter_matrix(corr_price, 
                               figsize=(7, 7), 
                               marker='x', 
                               alpha=0.4, 
                               color='black')
    plt.tick_params(axis = 'both', which = 'major', labelsize = 12)
    plt.tight_layout()
    plt.show()
```

```python
# –ì–¥–µ –º–æ–≥ - –Ω–∞–ø–∏—Å–∞–ª –ø–æ-–∞–Ω–≥–ª–∏–π—Å–∫–∏. –°–µ–π—á–∞—Å —É—Å—Ç–∞–ª. –ü–∏—à—É –ø–æ-—Ä—É—Å—Å–∫–∏. 
# –î–æ–±–∞–≤–ª—è–µ–º —Ü–∏—Ñ—Ä—É –≤–º–µ—Å—Ç–æ —Ä–µ–π—Ç–∏–Ω–≥–∞, —á—Ç–æ–±—ã —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é. 
def category_rating(data): 
    rating = data['rating']
    if rating == 'E':
        return 1
    elif rating == 'E10+':
        return 2
    elif rating =='M':
        return 3
    elif rating == 'T':
        return 4
    elif rating =='no_rating':
        return 5
    return 0
actual_df['category_rating'] = actual_df.apply(category_rating, axis=1)
```

```python
def user_region_portret(data,column):
    print('Most popular platform in '+column)
    sales_region_last_5_years(data,column)
    print('---------------------------------------------------------')
    print('')
    print('Most popular genre in '+column)
    genre_region = data.pivot_table(index='genre', 
                                    values=column, 
                                    aggfunc=['count','sum', 'median'])
    genre_region.columns = ['count_' + column, 'sum_' + column, 'median' + column]
    genre_region = genre_region.sort_values('sum_' + column, ascending=False).head()
    display(genre_region.style.background_gradient(cmap='Blues'))
    
    print('---------------------------------------------------------')
    
    genre_region['sum_' + column].plot(kind='bar',
                                       grid=True, 
                                       ylim=0, 
                                       legend=True,color='blueviolet')
    plt.xlabel('–ñ–∞–Ω—Ä –∏–≥—Ä—ã', fontsize=12)
    plt.title('–ü—Ä–æ–¥–∞–∂–∏ –ø–æ –∂–∞–Ω—Ä–∞–º', fontsize=14)
    plt.ylabel('–ú–∏—Ä–æ–≤—ã–µ –ø—Ä–æ–¥–∞–∂–∏, –º–ª–Ω. —à—Ç –≤ ' + column, fontsize=12)
    plt.show()
    
    print('---------------------------------------------------------')
    
    print('')
    print('Rating affection on '+column)
    
    rating_region = data.pivot_table(index='rating', 
                                     values=column, 
                                     aggfunc=['count','sum', 'median'])
    display(rating_region.style.background_gradient(cmap='Blues'))
    rating_region['sum'].plot(kind='bar',
                              grid=True, 
                              ylim=0, 
                              legend=True,
                              color='orange')
    plt.xlabel('–í–æ–∑—Ä–∞—Å—Ç–Ω–æ–π —Ä–µ–π—Ç–∏–Ω–≥ –∏–≥—Ä—ã', fontsize=12)
    plt.title('–ü—Ä–æ–¥–∞–∂–∏ –ø–æ —Ä–µ–π—Ç–∏–Ω–≥–∞–º', fontsize=14)
    plt.ylabel('–ú–∏—Ä–æ–≤—ã–µ –ø—Ä–æ–¥–∞–∂–∏, –º–ª–Ω. —à—Ç –≤ ' + column, fontsize=12)
    plt.show()
    cor_rating = actual_df['category_rating'].corr(actual_df[column])
    print('---------------------------------------------------------')
    print('Correlation between rating and',column,' {:.3f}'.format(cor_rating))
```

–ß—Ç–æ–±—ã —Å–æ–∑–¥–∞—Ç—å –ø–∞–ø–∫—É, –ø—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è —Ñ—É–Ω–∫—Ü–∏—è `mkdir`. –≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è —Å–º–æ–∂–µ—Ç —Å–æ–∑–¥–∞—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–Ω—É –ø–∞–ø–∫—É –≤ —Å—É—â–µ—Å—Ç–≤—É—é—â–µ–º –ø—É—Ç–∏, –µ—Å–ª–∏ —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –ø—É—Ç–∏ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç, —Ç–æ –≤—ã–∑–æ–≤–µ—Ç—Å—è –æ—à–∏–±–∫–∞. –ï—Å–ª–∏ –ø–∞–ø–∫–∞ —É–∂–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç - –≤—ã–∑–æ–≤–µ—Ç—Å—è –æ—à–∏–±–∫–∞

![](https://i.ibb.co/q53Z6YD/image.png)

```python
def mean_disctibution(column, a_group, b_group):
    a_rating = []
    b_rating = []
    df_a = actual_df[(actual_df[column] == a_group) & (~actual_df['user_score'].isna())]
    df_b = actual_df[(actual_df[column] ==b_group) &(~actual_df['user_score'].isna())]
    len_a = len(df_a)
    len_b = len(df_b)
    for _ in tqdm(range(10000)):
        sample_a = np.random.choice(df_a['user_score'], size=len_a, replace=True)
        sample_b = np.random.choice(df_b['user_score'], size=len_b, replace=True)
        a_rating.append(sample_a.mean())
        b_rating.append(sample_b.mean())
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15,5))
    ax.set_title('Group A (blueviolet) and B (orange) distribution of averages')
    ax.hist(a_rating, bins=100, color='blueviolet', alpha=0.5)
    ax.hist(b_rating, bins=100, color='orange', alpha=0.5)
    fig.tight_layout()
```

```python
def distribution_ttest(data, column, alpha, a_group, b_group):
    data = data[~data['user_score'].isna()]
    sample_a = data[data[column]==a_group]['user_score']
    sample_b = data[data[column]==b_group]['user_score']
    
    print('H0 - –¥–∞–Ω–Ω—ã–µ –ø–æ —Å—Ç–æ–ª–±—Ü—É', column,'—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ')
    print('H1 - –¥–∞–Ω–Ω—ã–µ –ø–æ —Å—Ç–æ–ª–±—Ü—É', column,'—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ')
    
    print('')
    x1,y1 = st.shapiro(sample_a)
    print('p-value', y1)
    if y1 < alpha: 
        print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ —Å—Ç–æ–ª–±—Ü—É user_score –∑–Ω–∞—á–µ–Ω–∏–π ' + a_group)
    else: 
        print('–ù–µ —É–¥–∞–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ —Å—Ç–æ–ª–±—Ü—É user_score –∑–Ω–∞—á–µ–Ω–∏–π ' + a_group)
    
    x2,y2 = st.shapiro(sample_b)
    print('')
    print('p-value', y2)
    if y2 < alpha: 
        print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ —Å—Ç–æ–ª–±—Ü—É user_score –∑–Ω–∞—á–µ–Ω–∏–π ' + b_group)
    else: 
        print('–ù–µ —É–¥–∞–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ —Å—Ç–æ–ª–±—Ü—É user_score –∑–Ω–∞—á–µ–Ω–∏–π ' + b_group)
        
    print('')
    print('H0 - –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫ –ø–æ —Å—Ç–æ–ª–±—Ü—É user_score —Ä–∞–≤–Ω—ã')
    print('H1 - –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫ –ø–æ —Å—Ç–æ–ª–±—Ü—É user_score –Ω–µ —Ä–∞–≤–Ω—ã')
    
    print('')
    results_levene = st.levene(sample_a, sample_b)
    print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results_levene.pvalue)
    if (results_levene.pvalue < alpha):
        print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫')
    else: 
        print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤—ã–±–æ—Ä–æ–∫')
    results = st.ttest_ind(sample_a,sample_b,equal_var = False)
    
    print('')
    print('H0 - —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ–∫ –ø–æ —Å—Ç–æ–ª–±—Ü—É user_score —Ä–∞–≤–Ω—ã')
    print('H1 - —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ–∫ –ø–æ —Å—Ç–æ–ª–±—Ü—É user_score –Ω–µ —Ä–∞–≤–Ω—ã')
    
    print('')
    print('p-value', results.pvalue)
    if results.pvalue < alpha: 
        print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ –¥–≤—É—Ö —Å—Ä–µ–¥–Ω–∏—Ö')
    else: 
        print('–ù–µ —É–¥–∞–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ –¥–≤—É—Ö —Å—Ä–µ–¥–Ω–∏—Ö')
```

### –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ;

```python
mean_disctibution('platform', 'PC', 'XOne')
distribution_ttest(actual_df, 'platform', 0.05, 'PC', 'XOne')
```
